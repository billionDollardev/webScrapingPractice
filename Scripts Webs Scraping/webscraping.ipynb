```python
lista_titulos=[]
lista_marcas=[]
lista_series=[]
lista_preciosR=[]
lista_preciosD=[]
lista_porcentajeD=[]
lista_urls=[]
```

```python
siguiente = ('https://www.mipcparquecentral.com/computadores/portatiles')
while True:
    r = requests.get(siguiente)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content, 'html.parser')
        titulos = soup.find_all('h3', attrs={"itemprop": "name"})
        titulos = [i.text for i in titulos]
        lista_titulos.extend(titulos)

        marcas = soup.find_all('span', attrs={"itemprop": "brand"})
        marcas = [i.text for i in marcas]
        lista_marcas.extend(marcas)

        series = soup.find_all('span', attrs={"class": "descripcion-referencia"})
        series = [i.text for i in series]
        lista_series.extend(series)

        dom = etree.HTML(str(soup))

        preciosR = dom.xpath('//div[@class="price-before"]//span')
        preciosR = [i.text for i in preciosR]
        lista_preciosR.extend(preciosR)

        preciosD = dom.xpath('//span[@itemprop="price"]')
        preciosD = [i.text for i in preciosD]
        lista_preciosD.extend(preciosD)

        porcentajeD = dom.xpath('//span[@class="tag_porcentaje tag_porcentaje_oferta"]')
        porcentajeD = [i.text for i in porcentajeD]
        lista_porcentajeD.extend(porcentajeD)

        links = dom.xpath('//div[@class="thumb-image"]//a')
        links = [i.get ('href') for i in links]
        lista_urls.extend(links)

        data_filled = [0 if x is None else x for x in lista_porcentajeD]
        data_filled = [0 if x is None else x for x in lista_preciosR]
        
        ini = soup.find('a', attrs={"href":"https://www.mipcparquecentral.com/computadores/portatiles?page=1&count=24"}).text
        ini = int(ini)
        can = soup.find('a', attrs={"href":"https://www.mipcparquecentral.com/computadores/portatiles?page=2&count=24"}).text
        
    else:
        break
    print(ini,can)
    
    if ini == can:
        break

    siguiente = dom.xpath('/html/body/section/div[2]/div[2]/div[2]/div[3]/div/div/ul/li[2]/a')[0].get('href')
```

```python
df1 = pd.DataFrame({"Titulos":lista_titulos, "Marcas":lista_marcas, "Series":lista_series,  "Precio Descuento":lista_preciosD, "Urls": lista_urls, "Urls":lista_urls})
df1
#df1.to_csv('data_portatil.csv')
```
